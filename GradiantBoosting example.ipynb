```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lr=0.5
```

```python
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
```

```python
f0=np.mean(y)
pred=np.full_like(y,f0,dtype=np.float64)
print(f"Initial prediction (f0):{f0}")
```

```python
r1=y-f0
print(f"Residual 1: {r1}")
```

```python
def tree1(x):
  return np.where(x<=2.5,-1,+1)

h1=tree1(x)
print(f"Tree 1: {h1}")
f1=f0+lr*h1
print(f"Prediction after first tree: {f1}")
```

```python
r2=y-f1
print(f"Residual 2: {r2}")
```

```python
def tree2(x):
  return np.where(x<=1.5,-1,np.where(x>=3.5,+1,0))

h2=tree2(x)
print(f"Tree 2: {h2}")
f2=f1+lr*h2
print(f"Prediction after second tree: {f2}")
```

```python
print("\nTrue y values:",y)
print("fInal prediction",f2)
print("Absolute error",np.abs(f2-y))
```
