```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
Iris=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
Iris['target']=iris.target
```

```python
Iris['flower_name']=Iris.target.apply(lambda x:iris.target_names[x])
Iris
```

```python
Iris0=Iris[:50]
Iris1=Iris[50:100]
Iris2=Iris[100:]
```

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(Iris0['sepal length (cm)'],Iris0['sepal width (cm)'],color='skyblue',marker='+')
plt.scatter(Iris1['sepal length (cm)'],Iris1['sepal width (cm)'],color='pink',marker='*')
plt.scatter(Iris2['sepal length (cm)'],Iris2['sepal width (cm)'],color='gray',marker='o')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(Iris0['petal length (cm)'],Iris0['petal width (cm)'],color='skyblue',marker='+')
plt.scatter(Iris1['petal length (cm)'],Iris1['petal width (cm)'],color='pink',marker='*')
plt.scatter(Iris2['petal length (cm)'],Iris2['petal width (cm)'],color='gray',marker='o')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=Iris.drop(['target','flower_name'],axis='columns')
```

```python
y=Iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model_c=SVC(C=10)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

```python
model_c=SVC(C=1)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

```python
model_linear_kernel=SVC(kernel='linear')
model_linear_kernel.fit(x_train,y_train)
```

```python
model_g=SVC(gamma=10)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```
