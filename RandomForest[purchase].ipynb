```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
purchase=pd.read_csv('/content/drive/MyDrive/Data Analytics/purchase.csv')
```

```python
purchase
```

```python
le=LabelEncoder()
purchase['Age']=le.fit_transform(purchase['Age'])
purchase['Gender']=le.fit_transform(purchase['Gender'])
purchase['Income']=le.fit_transform(purchase['Income'])
purchase['Education']=le.fit_transform(purchase['Education'])
purchase['Marital Status']=le.fit_transform(purchase['Marital Status'])
purchase['Purchased']=le.fit_transform(purchase['Purchased'])
```

```python
purchase
```

```python
purchase=purchase.drop(['ID'],axis=1)
```

```python
purchase
```

```python
independent_var=purchase[['Age','Gender','Income','Education','Marital Status']]
dependent_var=purchase['Purchased']
```

```python
independent_var
```

```python
dependent_var
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent_var,dependent_var,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy :{accuracy*100:.2f}%")
```
